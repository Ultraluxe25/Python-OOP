```python
class My:
    my_attr = 'lolo'
    def my_method(self):
        pass

print(My.__dict__)
```

```python
class My:
    x = 10
    y = 20

my_1 = My()
my_1.z = 30          # отдельно создадим атрибут экземпляра

print(My.__dict__)
```

```python
my_1.__dict__
```

```python
print(My.__dict__['x'])    # 10
print(My.__dict__['y'])    # 20
print(my_1.__dict__['z'])  # 30
```

### Есть простой способ, чтобы создать много атрибутов в экземпляре. Так как __dict__ - это словарь, то можно сразу присвоить ему словарь с именами атрибутов и значениями.

```python
class Animal:
    dict_all = {'name': 'Vasya', 'age': 15}

cat = Animal()
dog = Animal()

cat.__dict__ = Animal.dict_all    # __dict__ экземпляра cat ссылается на dict_all класса
dog.__dict__ = Animal.dict_all    # __dict__ экземпляра dog ссылается на dict_all класса

cat.name = 'Sonya'      # на самом деле name меняется в dict_all

print(cat.__dict__)     # {'name': 'Sonya', 'age': 15} 
print(dog.__dict__)     # {'name': 'Sonya', 'age': 15}
print(Animal.dict_all)  # {'name': 'Sonya', 'age': 15}
```

```python
class Student:
    def __init__(self, name, *args, **kwargs):
        self.name = name
        self.hobbies = args
        self.data = kwargs

# Создание экземпляра класса Student с дополнительными аргументами
student = Student("Вася", "Coding", "Teach", grade='5', school="Школа Магии")

print(student.name)     # Вася
print(student.hobbies)  # ('Coding', 'Teach')
print(student.data)     # {'grade': '5', 'school': 'Школа Магии'}
```

### Все действия прописанные в __init__ активируются только при создании экземпляра.
