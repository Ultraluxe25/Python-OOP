###   ,      .   __dict__ -  ,           .

```python
class Animal:
    dict_all = {'name': 'Vasya', 'age': 15}

cat = Animal()
dog = Animal()

cat.__dict__ = Animal.dict_all    # __dict__  cat   dict_all 
dog.__dict__ = Animal.dict_all    # __dict__  dog   dict_all 

cat.name = 'Sonya'      #    name   dict_all

print(cat.__dict__)     # {'name': 'Sonya', 'age': 15} 
print(dog.__dict__)     # {'name': 'Sonya', 'age': 15}
print(Animal.dict_all)  # {'name': 'Sonya', 'age': 15}
```
