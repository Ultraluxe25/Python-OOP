```python
class Person:
    def __init__(self, name):
        self.__name = name

    @property
    def name(self):            #   
        return self.__name

person = Person("Vasya")
print(person.name)  # Vasya
```

```python
class Person:
    def __init__(self, name):
        self.__name = name

    @property
    def name(self):            # 
        return self.__name

    @name.setter               # name -  ,   - "name"  
    def name(self, value):     #         - "name"
        self.__name = value

person = Person("John")
print(person.name) # John      #  

person.name = "Vasya"          #  
print(person.name) # Vasya
```

```python
class Person:
    def __init__(self, name):
        self.__name = name

    @property
    def name(self):                   #   
        return self.__name    

    @name.deleter
    def name(self):                   #   
        del self.__name

person = Person("john")
print(person.__dict__) # {'_Person__name': 'john'}
del person.name        #   
print(person.__dict__) # {}
```
